```python
x=data[['Channel', 'Region', 'Fresh', 'Milk', 'Grocery', 'Frozen','Detergents_Paper', 'Delicassen']]
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
  km = KMeans(n_clusters = i, n_init = 10)
  km.fit(x)
  css = km.inertia_
  wcss.append(css)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel('No of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#from elbow method 4 is the optimum number of clusters
```

```python
km = KMeans(n_clusters = 5, n_init = 10)
```

```python
km.fit(x)
```

```python
y_km = km.predict(x)
```

```python
y_km
```

```python
pd.Series(y_km).value_counts()
```

```python
data['Cluster'] = y_km
```

```python
data.head()
```

```python
colors = ['r','g','b','c','y']
for i in range(max(y_km)+1):
  cluster_data = x[y_km == i]
  plt.scatter(cluster_data['Grocery'],cluster_data['Fresh'],c=colors[i],label=f'Cluster{i}')
plt.xlabel('Grocery')
plt.ylabel('Fresh')
plt.title('Grocery vs Milk')
plt.legend()
```

```python
from sklearn.metrics import silhouette_score
silhoutte = silhouette_score(x,y_km)
print(silhoutte)
```

```python
#Agglomerative HC
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings = linkage(x,method = 'ward')
```

```python
plt.figure(figsize=(10,8))
dendrogram(mergings)
plt.axhline(35000,color='red',linestyle='--')
plt.grid()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 8, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
colors = ['r','g','b','c','y','m','k','w']
for i in range(max(y_hc)+1):
  cluster_data = x[y_hc == i]
  plt.scatter(cluster_data['Grocery'],cluster_data['Fresh'],c=colors[i],label=f'Cluster{i}')
plt.xlabel('Grocery')
plt.ylabel('Fresh')
plt.title('Grocery vs Milk')
plt.legend()
```

```python
from sklearn.metrics import silhouette_score
silhoutte = silhouette_score(x,y_hc)
print(silhoutte)
```
