```python
# Data Preprocessing
```

```python
import string
def remove_punc(tweet):
  punc_free = ''.join([i for i in tweet if i not in string.punctuation])
  return punc_free
```

```python
import nltk
nltk.download('punkt')
def tokenization(tweet):
  words = nltk.word_tokenize(tweet)
  return words
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(tweet):
  output = ''.join([i for i in tweet if i not in string.punctuation])
  return output
```

```python
from nltk.stem.wordnet import WordNetLemmatizer

nltk.download('wordnet')
wordnet_lem = WordNetLemmatizer()

def lemm(tweet):
  lemm_tweet = [wordnet_lem.lemmatize(word) for word in tweet]
  return lemm_tweet
```

```python
def preprocess(data_col):
  corpus = []
  for item in data_col:
    new_item = remove_punc(item)
    new_item = new_item.lower()
    new_item = tokenization(item)
    new_item = remove_stopwords(item)
    new_item = lemm(item)
    corpus.append(''.join(str(x)for x in new_item))
  return corpus
```

```python
corpus = preprocess(data['tweet'])
```

```python
corpus
```

```python
#Corpus
```

```python
##Bag-Of-Words
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(ngram_range=(1,2))
traindata = cv.fit_transform(corpus)
```

```python
x = traindata
y = data['label']
```

```python
x
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x,y)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rf.predict(x)
accuracy_score(y_pred,y)
```
